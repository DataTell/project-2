```python
# This's an example of how we can build our api to make a call for the user input of the zip code of interest and 
# a desired radius to find cats and dogs. The maximum radius can be 500 miles, but it may return a lot of pets.
# We can reduce the max radius to minimize the output and to satisfy the line limit on Heroku.
# Also, data comes back on more than 1 page. The max number of objects on 1 page can be 100, and there is has 
# to be an additional api call for each page. It requires to write a logic. If to use only one-page results, 
# we can end up having too little data because there can be cases when not many animals left after filtering all Nulls.
animal_type="dog"
animal_location="27560" # User inputs a zip code
animal_distance="50" # User inputs a radius 
animal_limit="100"
response_page="1"


animal_url="https://api.petfinder.com/v2/animals?type="+animal_type+"&location="+animal_location+"&distance="+animal_distance+"&limit="+animal_limit+"&page="+response_page
header={'Authorization': 'Bearer ' + accsess_token}

response=requests.get(animal_url, headers=header).json()
```
